```python
import plotly.express as px

# Loading the data
df = px.data.tips()

# Creating the bar chart
fig = px.bar(df, x='day', y="total_bill")

fig.show()
```

```python
import plotly.express as px

# Loading the iris dataset
df = px.data.tips()

fig = px.pie(df, values="total_bill", names="day")
fig.show()
```
